## 這堂課程採用 Google Colab

- Google Colab 全名為 Google Colaboratory，是由 Collaboration 與 Laboratory 兩個單字組合而成，直譯為「協作的實驗室」，意義是可以協同操作的科學計算平台。
- 前往 Google Colab 的首頁：<https://colab.google>。
- 點選 New Notebook 之後登入 Google 帳戶即可開始使用。

## 調整使用介面

- **說明 > 查看英文版本**。
- **Tools > Settings > Editor > Indentation width in spaces: 4**
- **Tools > Settings > Editor > Show line numbers**
- 如此一來，程式碼的字型是漂亮的、操作時跟課程錄影是契合的。

## Google Colab 很棒、很可愛的功能

- 可以切換不同的 runtime: **Runtime > Change runtime type**
- 可以添加程式碼儲存格亦可以添加文字儲存格。
- 可以切換科基模式、貓咪模式、螃蟹模式：**Settings > Miscellaneous**
